```python
import re
import pandas as pd
```

```python
filename = 'H2_HOFF_PUNIF_251016_130958.csv'

data = pd.read_csv(filename, header = 6)
```

```python
regex = re.compile(r'^.*(\-list|L(1|2))$')
columns_to_split = filter(regex.match, list(data.columns.values))
```

```python
num_values = 24
for c in columns_to_split:
    split_values = data[c].str.split(' ', num_values, expand = True)
    for i in range(1, len(split_values.columns)):
        data[c + '-' + str(i)] = split_values[i]
```
